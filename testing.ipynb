```python
import pandas as pd
```

```python
path = r"data\6g_testbed_dataset\EUR\6907619\split_output"

newPath = r"data\6g_testbed_dataset\EUR\6907619\newSplit"
```

```python
import os
```

```python
files = os.listdir(path)
files
```

```python
for f in files:
    currentTempFile = os.path.join(path, f)
    currentFileName = currentTempFile.split("\\")[-1].split(".")[0]
    newTempFile = os.path.join(newPath, currentFileName)
    print(currentFileName)
    tempDF = pd.read_csv(currentTempFile)
    tenmpCols = list(tempDF.columns)
    features = [col for col in tenmpCols if "lat" not in col]
    targets = [col for col in tenmpCols if "lat" in col]
    for target in targets:
        newTempDF = tempDF[features + [target]]
        newTempDF.to_csv(newTempFile+f"_{target}.csv", index=False)
    # print(tempDF.head())
    
```
